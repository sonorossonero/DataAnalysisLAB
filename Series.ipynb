```python
import numpy as np
import pandas as pd

# creating an series with ARRAY

data = np.array(['a','b','c','d'])
s = pd.Series(data)
print(s)

# creating a series with dict

d = {'asma' : 158, 'parveen' : 159}
s = pd.Series(d)
print(d)

# creating series with scalar values

s = (1,2,3,4,5)
indexx = list('abcde')

d = pd.Series(s, index = indexx)
print(d)
```

```python
# Indexing 

# Based on position

a = np.arange(5)
index = list('abcde')
s = pd.Series(a,index = index)
print(s)

print(s[:3]) # retrieves first 3 eles

print('------\n')

# Based on Label

print(s['e'])

# Fancy Indexing : It allows to use an array of indices to access multiple ele's at once

print(s[['a','b']]) 
```

```python
# Slicing : 

# Normal Slicing

print(s)
print(s[:-2]) # s[i:j-1]

#using label

print(s['a':'d']) # if we use labelled indexing it considers last one also in the output
```

```python
# Attributes
 
''' 
dtype
ndim
size
name
hasnans
index
empty

'''

fruits = np.array(['apple','mango','orange','grapes'])
indexx = list('abcd')
s = pd.Series(fruits, index = indexx, name = "FRUITS")
print(s)

print(s.dtype)
print(s.ndim)
print(s.size)
print(s.empty)
print(s.index)
print(s.name)
print(s.hasnans) # returns True or False if Nan is present
print(s.values,'\n')


# Methods

print(s.head(2))
print(s.tail(2))

print(s.count())
```

```python
# Mathematical operations on Series

''' 
The values with same indices are added otherwise return NaN
'''
s1= pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'])
s2 = pd.Series([10,20,-10,-50,100],index = ['z', 'y', 'a', 'c', 'e'])

# Addition of two Series
print("The series are: ",s1,s2)
print(s1+s2)
print("------\n")
print(s1.add(s2,fill_value = 0)) 
print(s1.subtract(s2,fill_value = 0))
print(s1.multiply(s2,fill_value = 0))
print(s1.div(s2,fill_value = 0))

```
